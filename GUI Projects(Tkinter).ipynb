```python
app = Tk()
app.title("Desktop APP")
app.geometry("350x350")
app.config(bg ='green')

e1 = Entry(app, width ="25", font=("Arial",15,"bold"), borderwidth="3")
e1.grid(columnspan=15, pady="15", padx="35")

calculation =""

def calculate(num):
    global calculation
    calculation+=str(num)
    e1.delete(0, "end")
    e1.insert(0, calculation)
    
    
def evaluate():
    global calculation
    try:
        calculation = str(eval(calculation))
        e1.delete(0, "end")
        e1.insert(0, calculation)
    except Exception:
        clear()
        #e1.insert(0, "Shey you dey whine mi ni!")
        e1.insert(0, "Math Error")
        
def clear():
    global calculation
    calculation=""
    e1.delete(0,"end")

b1 = Button(app, text="1", width="4",font=("Arial",15,"bold"), borderwidth="4", command=lambda:calculate(1))
b1.grid(row="1", column="0")


b2 = Button(app, text="2", width="4",font=("Arial",15,"bold"), borderwidth="4", command=lambda:calculate(2))
b2.grid(row="1", column="1")


b3 = Button(app, text="3", width="4",font=("Arial",15,"bold"), borderwidth="4", command=lambda:calculate(3))
b3.grid(row="1", column="2")


b4 = Button(app, text="+", width="4",font=("Arial",15,"bold"), borderwidth="4", command=lambda:calculate("+"))
b4.grid(row="1", column="3")


b5 = Button(app, text="4", width="4",font=("Arial",15,"bold"), borderwidth="4", command=lambda:calculate(4))
b5.grid(row="2", column="0",pady="5")


b6 = Button(app, text="5", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(5))
b6.grid(row="2", column="1")


b7 = Button(app, text="6", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(6))
b7.grid(row="2", column="2")


b8 = Button(app, text="-", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate("-"))
b8.grid(row="2", column="3")


b9 = Button(app, text="7", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(7))
b9.grid(row="3", column="0")


b10 = Button(app, text="8", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(8))
b10.grid(row="3", column="1")


b11 = Button(app, text="9", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(9))
b11.grid(row="3", column="2")


b12 = Button(app, text="x", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate("*"))
b12.grid(row="3", column="3")


b13 = Button(app, text="0", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(0))
b13.grid(row="4", column="0")


b14 = Button(app, text="(", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate("("))
b14.grid(row="4", column="1")


b15 = Button(app, text=")", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate(")"))
b15.grid(row="4", column="2")


b16 = Button(app, text="/", width="4",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate("/"))
b16.grid(row="4", column="3")


b17 = Button(app, text="C", width="5",font=("Arial",15,"bold"), borderwidth="4", command=clear)
b17.grid(row="5", column="0")


b18 = Button(app, text="=", width="5",font=("Arial",15,"bold"), borderwidth="4",command=evaluate)
b18.grid(row="5", column="1")


b19 = Button(app, text="^", width="5",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate("**"))
b19.grid(row="5", column="2")

b20 = Button(app, text=".", width="5",font=("Arial",15,"bold"), borderwidth="4",command=lambda:calculate("."))
b20.grid(row="5", column="3")



app.mainloop()
```

```python
app = Tk()
app.title("Body Mass Index")
app.geometry("400x300")
app.config(bg="cyan")


class NewWindow(Toplevel):
    def __init__(self):
        super().__init__()
        self.title("BMI Calc")
        self.geometry("500x400")
        self.config(bg="purple")
            
        self.title1 = Label(self, text="BMI Calculator",font=('arial',20), width=15)
        self.title1.place(x = 80, y = 10)
        
        
        self.user = Label(self, text="Your Weight", font=('arial',15), width=12)
        self.user.place(x = 10, y = 80)

        self.password = Label(self, text="Your Height", font=('arial',15), width=12)
        self.password.place(x = 10, y = 120)
    
        
        self.wt = Entry(self, width=17,font=('arial',15))
        self.wt.place(x =180, y = 80)

        self.ht = Entry(self, width=17,font=('arial',15))
        self.ht.place(x =180, y = 120)


        self.result = Label(self, text="", font=('arial',15), width=42)
        self.result.place(x = 10, y = 300)
        
        self.comment = Label(self, text="", font=('arial',15), width=42)
        self.comment.place(x = 10, y = 340)
        
        
        self.button = Button(self, text="COMPUTE", width=10,font=('arial',15, "bold"), bd=5, command= self.compute)
        self.button.place(x = 120, y = 170)
        
    def compute(self):
        self.w = float(self.wt.get())
        self.h = float(self.ht.get())
        self.bmi = self.w/(self.h*self.h)
        self.bmi=round(self.bmi,4)
        self.result.config(text=f"Your BMI is {self.bmi}")
        
        if self.bmi<18.5:
            self.comment.config(text="You are underweight, please eat very well.")
        elif self.bmi>18.5 and self.bmi<24.9:
            self.comment.config(text="Your health is in good condition, please keep it up!.")
        elif self.bmi>24.9 and self.bmi<=29.9:
            self.comment.config(text="You are overweight, please watch your weight.")
        else:
            self.comment.config(text="Your health is in a critical condition, please visit your doctor immediately!.")
            
            
            
    
        
        

        
title = Label(app, text="Login Page", bg="pink", font=('arial',20), width=17)
title.place(x = 80, y = 10)

luser = Label(app, text="User Name", font=('arial',15), width=12)
luser.place(x = 30, y = 60)

lpass = Label(app, text="Password", font=('arial',15), width=12)
lpass.place(x = 30, y = 110)


euser = Entry(app, width=17,font=('arial',15))
euser.place(x =200, y = 60)

epass = Entry(app, width=17,font=('arial',15),show="*")
epass.place(x =200, y = 110)



def new_app(user, password):
    global new_window
    
    if user==euser.get() and password==epass.get():
        new_window=NewWindow()
    else:
        messagebox.showerror("Error","User credentials are not correct")
        
        
        
        
def clear():
    euser.delete(0,END)
    epass.delete(0,END)
    

add = Button(app, text="BMI App", width=10,font=('arial',15, "bold"), bd=5, command=lambda:new_app("admin", "123") )
add.place(x = 30, y = 150)


clear = Button(app, text="CLEAR", width=10,font=('arial',15, "bold"), bd=5, command=clear )
clear.place(x =220 , y = 150)








app.mainloop()
```
